```python
from lattice import WSe
from minimize_energy import *
import numpy as np
```

```python
l = 20
x0 = 1.1
k = 1
L = WSe(l, k, x0=x0, periodic=True)
```

```python
%matplotlib notebook
L.plot_lattice(all=True)
#idx = np.random.randint(2*L.N)
idx = L.ij_to_idx(int(l/2)+1,int(l/2))

nb = L.neighbors(idx)
print(nb)
print([L.idx_to_ij(n) for n in nb])
L.plot_points(nb, 'b')
L.plot_points([idx], 'g')
plt.show()
print(L.energy())
```

```python
%matplotlib notebook
idx = L.ij_to_idx(int(l/2),int(l/2))
L.remove_atom(idx+L.N)
L.plot_lattice(all=True)
plt.show()
print(L.energy())
```
